Парсер товаров с любого раздела веб-сайта Агросервер.ру
В качестве примера буду рассматривать страницу с семенами расторопши
https://agroserver.ru/rastoropsha/

Описание: Этот скрипт представляет собой простой парсер веб-страницы, который извлекает информацию о товарах с указанной страницы и выводит её в консоль.

Использование:

Установите Python (если не установлен) и необходимые библиотеки: requests и BeautifulSoup.
Запустите скрипт в интерпретаторе Python.
Введите URL страницы с товарами (раздела Агросервер) в переменную url.
Запустите скрипт.
Скрипт извлечет информацию о каждом товаре на странице и выведет её в консоль.
Функциональность:

Извлечение даты публикации объявления о товаре.
Извлечение города, указанного продавцом товара.
Извлечение цены товара.
Извлечение имени продавца.
Извлечение контактного телефона продавца.
Извлечение рейтинга продавца.
Зависимости:

Python 3.x
Библиотека requests
Библиотека BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup

# URL страницы с товарами
url = "https://agroserver.ru/rastoropsha/"

# Заголовки для HTTP-запроса
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Выполнение HTTP-запроса
response = requests.get(url, headers=headers)

# Проверка статуса ответа
if response.status_code == 200:
    # Создание объекта BeautifulSoup для парсинга HTML
    soup = BeautifulSoup(response.content, 'html.parser')

    # Поиск всех элементов с товарами (предполагая, что товары находятся в элементах <div> с классом 'wrapper')
    items = soup.find_all('div', class_='wrapper')

    # Перебор всех найденных товаров и извлечение информации
    for item in items:
        # Извлечение даты
        date_elem = item.find('div', class_='date')
        date = date_elem.get_text(strip=True) if date_elem else 'Не указано'

        # Извлечение города
        city_elem = item.find('div', class_='geo')
        city = city_elem.get_text(strip=True) if city_elem else 'Не указано'

        # Извлечение цены
        price_elem = item.find('div', class_='price')
        price = price_elem.get_text(strip=True) if price_elem else 'Не указано'

        # Извлечение имени
        name_elem = item.find('a', class_='personal_org_menu')
        name = name_elem.get_text(strip=True) if name_elem else 'Не указано'

        # Извлечение телефона
        phone_elem = item.find('div', class_='phone2')
        phone = phone_elem.find('a').get_text(strip=True) if phone_elem and phone_elem.find('a') else 'Не указано'

        # Извлечение рейтинга
        rating_elem = item.find('div', class_='rating')
        rating = rating_elem.find('a').get_text(strip=True) if rating_elem and rating_elem.find('a') else 'Не указано'

        # Вывод информации о товаре
        print(f"Дата: {date}")
        print(f"Город: {city}")
        print(f"Цена: {price}")
        print(f"Имя: {name}")
        print(f"Телефон: {phone}")
        print(f"Рейтинг: {rating}")
        print("-" * 40)
else:
    print("Не удалось загрузить страницу")
```
